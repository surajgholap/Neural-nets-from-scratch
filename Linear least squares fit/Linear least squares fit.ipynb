```python
def plot_func(w,y):
    a = float(w[0][0])
    b = float(w[1])
    plt.figure(figsize=(10,10))
    x1 = np.arange(1,51)
    plt.scatter(x1,y, c = 'green')
    plt.plot(x1,a*x1+b, c = 'red')
    plt.ylabel('Y values')
    plt.xlabel('X values')
    plt.title('')
    plt.show()
```

```python
def gradient(x,y,w0,w1):
  a = 2*(w1*x + w0 - y)
  b = 2*x*(w1*x + w0 - y)
  return np.array([a,b])
```

```python
def grad_descent(w,x,y):
  w0 = np.float(w[0])
  w1 = np.float(w[1])
  for i in range(50):
        while True:
            g = gradient(x[i],y[i],w[0],w[1])
            update = learning_rate * g
            w_n = np.subtract(w,update)
            if abs(np.linalg.norm(w - w_n)) < threshold:
                return w
            w =  w_n
  return w
```

```python
x = np.arange(1,51) 
y = []
for i in x:
  a = i + np.random.uniform(-1,1)
  y.append(a)
D = np.asarray(y)
D.resize((1,50))
ones = [1 for i in range(50)]
X = np.vstack([x, ones])
```

```python
weights = exact_sol(X,D)
plot_func(weights, y)
```
